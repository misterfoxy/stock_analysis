```python
# import dependencies
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import requests
from config import api_key
```

```python
path_to_load = "nyse_list.csv"
```

```python
# load the data
stock_list_df = pd.read_csv(path_to_load)
stock_list_df
```

```python
base_url="https://cloud.iexapis.com/stable/stock/market/batch?token="+ api_key

stock_data = []

print('-----')
print('Initiaiting Retrieval')
print('-----')
for i,stock in enumerate(stock_list_df["ACT Symbol"]):
    
    stock_url = base_url + "&symbols="+stock+"&types=quote,earnings,estimates,balance-sheet"
    
    try:
        response = requests.get(stock_url, timeout=260).json()
#         print(stock, response[stock]["quote"]["latestPrice"])
        stock_price = response[stock]["quote"]["latestPrice"]
        stock_pe = response[stock]["quote"]["peRatio"]
        stock_change_price = response[stock]["quote"]["change"]
        stock_change_percent = response[stock]["quote"]["changePercent"]
        stock_ytdChange = response[stock]["quote"]["ytdChange"]
        stock_consensus_EPS = response[stock]["earnings"]["consensusEPS"]
        stock_actual_EPS = response[stock]["earnings"]["actualEPS"]
        
        stock_data.append({"Stock": stock,
                           "Price": stock_price,
                           "PE": stock_pe,
                           "Change":stock_change_price,
                           "Change %": stock_change_percent,
                           "YTD Change": stock_ytdChange,
                           "Consensus EPS": stock_consensus_EPS,
                           "Actual EPS": stock_actual_EPS})
        
    except:
        print('Error fetching data')
        pass
print('-----')
print('Retrieval Complete')
print('-----')
```
